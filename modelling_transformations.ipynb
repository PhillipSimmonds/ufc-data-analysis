```python
import pandas as pd
import datetime
import numpy as np
```

```python
edf = pd.read_csv("extraction/events.csv")
edf = edf.drop("Unnamed: 0",axis=1)
edf
```

```python
df = pd.read_csv("extraction/fight_records.csv")
df = df.drop("Unnamed: 0",axis=1)
df = pd.merge(df,edf, how = "left", left_on = df.eventurlid, right_on= edf.eventurlid)
df = df.drop(["eventurlid_y", "eventurl","fighttitlefullname"],axis=1)
df
```

```python
# Change date format
def change_date(d):
    conv = {
        "January": "01",
        "February": "02",
        "March": "03",
        "April": "04",
        "May": "05",
        "June": "06",
        "July": "07",
        "August": "08",
        "September": "09",
        "October": "10",
        "November": "11",
        "December": "12",
    }
    return [int(d.split(" ")[2]),
            int(conv[d.split(" ")[0]]),
            int(d.split(" ")[1].replace(",",""))]

date = [datetime.date(change_date(x)[0],change_date(x)[1],change_date(x)[2]) for x in df.datestring]
df["date"] = date
df = df.loc[df.strikes != "--"]
```
